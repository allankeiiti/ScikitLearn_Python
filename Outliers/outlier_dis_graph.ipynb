## Detectando outliers na base credit_data.csv com Gráfico de Dispersão

```python
import matplotlib.pyplot as plt
import pandas as pd
base = pd.read_csv('credit_data.csv')
base = base.dropna()
```

```python
# Income x Age - há outliers
plt.scatter(base.iloc[:, 1], base.iloc[:, 2])
plt.show()
```

```python
# Income x Loan - Não há outliers
plt.scatter(base.iloc[:, 1], base.iloc[:, 3])
plt.show()
```

```python
# Age x Loan - há outliers
plt.scatter(base.iloc[:, 2], base.iloc[:, 3])
plt.show()
```
