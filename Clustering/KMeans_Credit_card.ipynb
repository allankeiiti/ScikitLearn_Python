### Clustering com KMeans utilizando base credit_card_clients

```python
import pandas as pd
base = pd.read_csv('credit_card_clients.csv', header=1)
```

```python
base.head()
```

### Importando módulos

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Criando novo atributo de BILL_TOTAL
base['BILL_TOTAL'] = base['BILL_AMT1'] + base['BILL_AMT2'] + base['BILL_AMT3'] + base['BILL_AMT4'] + base['BILL_AMT5'] + \
 base['BILL_AMT6']
```

```python
# Detalhe que foi utilizado os atributos: LIMIT_BAL(1) e BILL_TOTAL(25) 
X = base.iloc[:, [1, 25]].values
X
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

### Descobrindo quantos clusters devem ser inseridos no K com o método <i>Elbow</i>

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
wcss
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 11), wcss)
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

### Determinando quantos Clusters utilizar 
#### Fonte do método: https://medium.com/pizzadedados/kmeans-e-metodo-do-cotovelo-94ded9fdf3a9

```python
from math import sqrt
def optimal_number_of_clusters(wcss):
    x1, y1 = 2, wcss[0]
    x2, y2 = 20, wcss[len(wcss)-1]

    distances = []
    for i in range(len(wcss)):
        x0 = i+2
        y0 = wcss[i]
        numerator = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominator = sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distances.append(numerator/denominator)
    
    return distances.index(max(distances)) + 2
print(f'Será utilizado {optimal_number_of_clusters(wcss)} clusters.')
```

```python
kmeans = KMeans(n_clusters=6, random_state=0)
previsoes = kmeans.fit_predict(X)
```

```python
plt.scatter(X[previsoes == 0, 0], X[previsoes == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[previsoes == 1, 0], X[previsoes == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[previsoes == 2, 0], X[previsoes == 2, 1], s=100, c='orange', label='Cluster 3')
plt.scatter(X[previsoes == 3, 0], X[previsoes == 3, 1], s=100, c='green', label='Cluster 4')
plt.scatter(X[previsoes == 4, 0], X[previsoes == 4, 1], s=100, c='gray', label='Cluster 5')
plt.scatter(X[previsoes == 5, 0], X[previsoes == 5, 1], s=100, c='purple', label='Cluster 6')
plt.xlabel('Limite')
plt.ylabel('Gastos')
plt.legend()
plt.show()
```
